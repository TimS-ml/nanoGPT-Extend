```python
from diffusers import DDPMPipeline, DDIMPipeline
from IPython.display import display, display_html
from boring_utils.utils import get_device, cprint, tprint

model_id = "google/ddpm-celebahq-256"
device = get_device()
cprint(device)
```

```python
# load model and scheduler
pipe = DDPMPipeline.from_pretrained(model_id, num_inference_steps=1000)  # you can replace DDPMPipeline with DDIMPipeline or PNDMPipeline for faster inference
# pipe = DDIMPipeline.from_pretrained(model_id, num_inference_steps=100)
pipe.to(device)
```

```python
# run pipeline in inference (sample random noise and denoise)
image = pipe().images
```
